### 1. Monthly Sales Trends & Highest Revenue Month

```python
# Monthly revenue
monthly = df.resample("M", on="Date")["Total_Sales_INR"].sum().reset_index()

# Highest revenue month
best_month = monthly.loc[monthly["Total_Sales_INR"].idxmax()]

print(monthly)
print("\nHighest Revenue Month:\n", best_month)

# Visual
plt.plot(monthly["Date"], monthly["Total_Sales_INR"])
plt.xticks(rotation=45)
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.tight_layout()
plt.show()
```

#### 2. Which product categories generate the most revenue and have the highest number of orders?

```python
# ---- Revenue per category ----
cat_rev = df.groupby('Product_Category')['Total_Sales_INR'].sum()
cat_rev = cat_rev.sort_values(ascending=False)

# ---- Number of orders per category ----
cat_orders = df['Product_Category'].value_counts()

# ---- Plot side-by-side ----
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))

# Revenue
cat_rev.head(8).plot(kind='bar', ax=ax1, color='skyblue')
ax1.set_title('Top Categories â€“ Revenue')
ax1.set_ylabel('Revenue (INR)')
ax1.tick_params(axis='x', rotation=45)

# Orders
cat_orders.head(8).plot(kind='bar', ax=ax2, color='lightgreen')
ax2.set_title('Top Categories â€“ Orders')
ax2.set_ylabel('Number of Orders')
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# ---- Print the winners ----
print("Highest revenue category :", cat_rev.index[0], f"â‚¹{cat_rev.iloc[0]:,.0f}")
print("Most ordered category   :", cat_orders.index[0], f"{cat_orders.iloc[0]} orders")
```

#### 3. Who are the top customers by total spending?

```python
# Step 1: Add up all money spent by each customer
customer_money = df.groupby('Customer_ID')['Total_Sales_INR'].sum()

# Step 2: Sort from highest to lowest and pick top 5
top_5 = customer_money.sort_values(ascending=False).head(5)

# Step 3: Make a nice bar chart
top_5.plot(kind='bar', color='gold', figsize=(8,5))
plt.title('Top 5 Customers by Spending')
plt.ylabel('Total Spent (INR)')
plt.xlabel('Customer ID')
plt.xticks(rotation=0)  # keep names straight
plt.grid(axis='y', alpha=0.3)
plt.show()

# Step 4: Print the #1 customer
best_cust = top_5.index[0]
best_amount = top_5.iloc[0]
print(f"Top customer: {best_cust} â†’ â‚¹{best_amount:,.0f}")
```

#### 4. Which states have the highest sales?

```python
# Step 1: Add up sales for each state
state_sales = df.groupby('State')['Total_Sales_INR'].sum()

# Step 2: Sort from highest to lowest, take top 7
top_states = state_sales.sort_values(ascending=False).head(7)

# Step 3: Draw a horizontal bar chart (easier to read long names)
top_states.plot(kind='barh', color='teal', figsize=(8,5))
plt.title('Top 7 States by Sales')
plt.xlabel('Total Sales (INR)')
plt.ylabel('State')
plt.grid(axis='x', alpha=0.3)
plt.show()

# Step 4: Print the winner
winner = top_states.index[0]
amount = top_states.iloc[0]
print(f"Top state: {winner} â†’ â‚¹{amount:,.0f}")
```

#### 5. Which payment methods are used most and make the most money

```python
# Step 1: Count how many times each payment method is used
payment_count = df['Payment_Method'].value_counts()

# Step 2: Add up total sales for each payment method
payment_money = df.groupby('Payment_Method')['Total_Sales_INR'].sum()

# Step 3: Sort money from highest to lowest
payment_money = payment_money.sort_values(ascending=False)

# Step 4: Make two charts
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))

# Left: Pie chart for "how many people use it"
payment_count.plot(kind='pie', ax=ax1, autopct='%1.0f%%', startangle=90)
ax1.set_title('Payment Methods Used')
ax1.set_ylabel('')  # remove label

# Right: Bar chart for "how much money it brings"
payment_money.plot(kind='bar', ax=ax2, color='purple')
ax2.set_title('Revenue by Payment Method')
ax2.set_ylabel('Total Sales (INR)')
ax2.tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()

# Step 5: Print top results
print("Most used payment :", payment_count.index[0], f"({payment_count.iloc[0]} orders)")
print("Highest revenue   :", payment_money.index[0], f"â†’ â‚¹{payment_money.iloc[0]:,.0f}")
```

#### 6. What % of orders are Delivered vs Returned? Which categories have high returns?

```python
# === PART 1: Pie Chart - Delivered vs Returned ===
status_count = df['Delivery_Status'].value_counts()

# Use 2 clear colors: Green = Delivered, Red = Returned
colors = ['lightgreen', 'lightcoral']

plt.figure(figsize=(6, 6))
status_count.plot(kind='pie', 
                  autopct='%1.1f%%', 
                  colors=colors, 
                  startangle=90)
plt.title('Order Status: Delivered vs Returned', fontsize=14, fontweight='bold')
plt.ylabel('')  # Remove extra label
plt.show()

# === PART 2: Bar Chart - Which Categories Have Most Returns? ===
# Keep only Delivered & Returned orders
orders = df[df['Delivery_Status'].isin(['Delivered', 'Returned'])]

# Count total orders per category
total_orders = orders['Product_Category'].value_counts()

# Count returned orders per category
returned_orders = orders[orders['Delivery_Status'] == 'Returned']['Product_Category'].value_counts()

# Calculate return % = (returned / total) * 100
return_percent = (returned_orders / total_orders * 100).round(1)

# Sort and take top 6
top_returns = return_percent.sort_values(ascending=False).head(6)

# Plot
plt.figure(figsize=(8, 5))
top_returns.plot(kind='bar', color='salmon', edgecolor='black')
plt.title('Top 6 Categories with Highest Return Rate', fontsize=14, fontweight='bold')
plt.ylabel('Return Rate (%)')
plt.xlabel('Product Category')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# === PART 3: Print Key Results ===
overall_return = (len(orders[orders['Delivery_Status'] == 'Returned']) / len(orders)) * 100

print(f"Overall Return Rate: {overall_return:.1f}%")
print(f"Worst Category: {top_returns.index[0]} â†’ {top_returns.iloc[0]:.1f}% returns")
```

#### 7. How do sales change over the days of the week? Which day is best?

```python
# === PART 1: Add Day Name (Monday, Tuesday...) ===
df['Day_Name'] = df['Date'].dt.day_name()

# === PART 2: Total Sales by Day ===
sales_by_day = df.groupby('Day_Name')['Total_Sales_INR'].sum()

# Put days in order: Mon â†’ Sun
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sales_by_day = sales_by_day.reindex(day_order)

# === PART 3: Plot a Simple Bar Chart ===
plt.figure(figsize=(9, 5))
sales_by_day.plot(kind='bar', color='skyblue', edgecolor='navy')

plt.title('Total Sales by Day of the Week', fontsize=14, fontweight='bold')
plt.ylabel('Sales (INR)')
plt.xlabel('Day')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# === PART 4: Find & Print the Best Day ===
best_day = sales_by_day.idxmax()
best_amount = sales_by_day.max()

print(f"Best sales day: {best_day} â†’ â‚¹{best_amount:,.0f}")
```

#### 8. Which products are the most profitable?

```python
# Top 5 Products by Sales (if no Cost)

# Group by Product Name and sum sales
sales_by_product = df.groupby('Product_Name')['Total_Sales_INR'].sum()

# Top 5
top_5 = sales_by_product.sort_values(ascending=False).head(5)

# Plot
plt.figure(figsize=(9, 5))
top_5.plot(kind='bar', color='lightblue', edgecolor='navy')
plt.title('Top 5 Products by Sales Revenue', fontsize=14, fontweight='bold')
plt.ylabel('Total Sales (INR)')
plt.xlabel('Product')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Print #1
print(f"Top Product: {top_5.index[0]} â†’ â‚¹{top_5.iloc[0]:,.0f}")
```

#### 9. Which customers buy the most items?

```python
# === PART 1: Add up total quantity bought by each customer ===
qty_by_cust = df.groupby('Customer_ID')['Quantity'].sum()

# === PART 2: Sort and take top 5 ===
top_5_qty = qty_by_cust.sort_values(ascending=False).head(5)

# === PART 3: Plot a clean bar chart ===
plt.figure(figsize=(9, 5))
top_5_qty.plot(kind='bar', color='orange', edgecolor='darkorange')

plt.title('Top 5 Customers by Quantity Purchased', fontsize=14, fontweight='bold')
plt.ylabel('Total Items Bought')
plt.xlabel('Customer ID')
plt.xticks(rotation=0)  # Keep IDs readable
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# === PART 4: Print the #1 buyer ===
top_buyer = top_5_qty.index[0]
top_items = top_5_qty.iloc[0]

print(f"Biggest Buyer: {top_buyer}")
print(f"Total Items: {top_items:,}")
```

### Final Dashboard: All Key Insights in One View!

```python
# === Create 2x2 Dashboard ===
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('AMAZON SALES 2025 â€“ QUICK DASHBOARD', fontsize=18, fontweight='bold')

# === 1. Monthly Sales Trend ===
df['Month'] = df['Date'].dt.strftime('%Y-%m')
monthly = df.groupby('Month')['Total_Sales_INR'].sum()
monthly.plot(ax=axes[0,0], kind='line', marker='o', color='blue')
axes[0,0].set_title('Monthly Sales Trend')
axes[0,0].set_ylabel('Sales (INR)')
axes[0,0].tick_params(axis='x', rotation=45)
axes[0,0].grid(True, alpha=0.3)

# === 2. Top 5 Categories by Revenue ===
cat_rev = df.groupby('Product_Category')['Total_Sales_INR'].sum().sort_values(ascending=False).head(5)
cat_rev.plot(ax=axes[0,1], kind='bar', color='green')
axes[0,1].set_title('Top 5 Categories (Revenue)')
axes[0,1].set_ylabel('Revenue')
axes[0,1].tick_params(axis='x', rotation=45)

# === 3. Top 5 States ===
state_sales = df.groupby('State')['Total_Sales_INR'].sum().sort_values(ascending=False).head(5)
state_sales.plot(ax=axes[1,0], kind='barh', color='purple')
axes[1,0].set_title('Top 5 States')
axes[1,0].set_xlabel('Sales (INR)')

# === 4. Payment Method Pie ===
payment = df['Payment_Method'].value_counts()
payment.plot(ax=axes[1,1], kind='pie', autopct='%1.0f%%', colors=['gold','lightblue','lightgreen','pink'])
axes[1,1].set_title('Payment Methods Used')
axes[1,1].set_ylabel('')

# === Final Touch ===
plt.tight_layout()
plt.subplots_adjust(top=0.93)
plt.show()

# === Print Summary ===
print("DASHBOARD SUMMARY")
print(f"â€¢ Total Sales: â‚¹{df['Total_Sales_INR'].sum():,.0f}")
print(f"â€¢ Total Orders: {len(df):,}")
print(f"â€¢ Best Month: {monthly.idxmax()}")
print(f"â€¢ Top Category: {cat_rev.index[0]}")
print(f"â€¢ Top State: {state_sales.index[0]}")
```
